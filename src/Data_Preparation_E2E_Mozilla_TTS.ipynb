```
####Processing content - Set 1
#input:
#transcript.txt: FPT Open Speech Data transcript
#junks.txt: contains all junk file names (excluding extension), these files will not be included in train/test/validate data
#output: metadata.txt
#content: path|text

import os, sys
from datetime import datetime

t1 = datetime.now()

# Open a file
dataset = "FPTOpenSpeechData"

txt_src = "/content/data-prep-e2e-mozilla-tts/data/FPTOpenSpeechData_Set001_V0.1/transcript.txt"
#txt_prefix = "/content/data-prep-e2e-mozilla-tts/data/FPTOpenSpeechData_Set001_V0.1/wav/"
txt_prefix = ''

fmeta = "/content/data-prep-e2e-mozilla-tts/data/FPTOpenSpeechData_Set001_V0.1/metadata.csv"
fjunks = "/content/data-prep-e2e-mozilla-tts/data/junks.txt"

if os.path.isfile(txt_src):
    with open(txt_src, 'r', encoding="utf-8") as myfile:
        while True:
            isjunk = 0
            data = myfile.readline()
            if not data:
                # eof
                break
                
            with open(fjunks, 'r', encoding="utf-8") as f:
                while True:
                    datajunk = f.readline()
                    if not datajunk:
                        # eof
                        break
                    #print(data)
                    if (data.find(datajunk) != -1) :
                        isjunk = 1
                        print('junk = ', datajunk)
                        
            if (isjunk == 0):
                x = data.split('|')         #separate string
                #print(x)
                x2 = x[0].split(".mp3")     #separate .mp3
                data2write = txt_prefix + x2[0] + "|" + x[1] + "|" + x[1] + "\n"
                #print(data2write)
          
                with open(fmeta, "a", encoding="utf-8") as myfile2:
                    myfile2.write(data2write)

t2 = datetime.now()
print(t2-t1)
```
